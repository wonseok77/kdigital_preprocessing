```python
kto_201901['국적'].unique()
```

```python
continents_list=['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙',  '교포소계']
continents_list
```

```python
# 국적 데이터에 continents_list에 대륙 이름이 없는 것만 필터링하기..
kto_201901_country = kto_201901[~kto_201901['국적'].isin(continents_list)]
kto_201901_country.head()
```

```python
kto_201901_country['국적'].unique()
```

```python
kto_201901_country_newindex = kto_201901_country.reset_index(drop=True)
kto_201901_country_newindex.head()
```

```python
# 각 국가별 국내에 관광을 위해 입국하는 관광객비욜(%) 칼럼 생성하기
# 관광객비율(%) = 관광/계 * 100 (소숫점 첫째자리까지)
kto_201901_country_newindex['관광객비율(%)'] = round((kto_201901_country_newindex['관광'] / \
                                                      kto_201901_country_newindex['계']) * 100,1)
kto_201901_country_newindex.head()
```

```python
kto_201901_country_newindex.sort_values(by = '관광객비율(%)',ascending=False).head()
```

```python
# 컬럼명 : 전체비율(%)
kto_201901_country_newindex['전체비율(%)'] = round((kto_201901_country_newindex['관광'] /\
                                                   kto_201901_country_newindex['관광'].sum()) * 100, 1)
kto_201901_country_newindex.head()
```

```python
# 전체 116개 통합하기
# 통합에 사용할 데이터프레임 변수명 : df
# 전체 통합 후 조회하기
df = pd.DataFrame()

for yyyy in range(2010, 2020) :
    for mm in range(1,13) :
        try :
            file_path = './files2/kto_{}{}.xlsx'.format(yyyy,str(mm).zfill(2))
            df2 = pd.read_excel(file_path, header = 1 , skipfooter = 4, usecols="A:G")
            df2['기준년월'] = str(yyyy) + '-' + str(mm).zfill(2) 
            df2 = df2[~df2['국적'].isin(continents_list)]
            df2['대륙'] = continents
            df2['관광객비율(%)'] = round((df2['관광'] / df2['계']) * 100,1)
            df2['전체비율(%)'] = round((df2['관광'] / df2['관광'].sum()) * 100, 1)


            df = pd.concat([df, df2], axis=0,ignore_index=True)
        except :
            pass
df
```

```python
# 강사풀이
# shift + tab
df = pd.DataFrame()

for y in range(2010, 2020, 1) :
    for m in range(1, 13, 1) :
        try :
            # 처리 필요
            file_nm = str(y) + str(m).zfill(2)
            file_path = "./files2/kto_{}.xlsx".format(file_nm)
            kto_201901 = pd.read_excel(file_path, header=1, skipfooter=4, usecols="A:G")

            # 처리 필요
            kto_201901["기준년월"] = str(y) + "-" + str(m).zfill(2)

            continents_list = ['아시아주', '미주', '구주', '대양주', 
                               '아프리카주', '기타대륙', '교포소계']
            condition = kto_201901["국적"].isin(continents_list)
            kto_201901_country = kto_201901[condition == False]

            kto_201901_country_newindex = kto_201901_country.reset_index(drop=True)

            continents = ["아시아"] * 25 + ["아메리카"] * 5 + ["유럽"] * 23 + \
                            ["오세아니아"] * 3 + ["아프리카"] * 2 + ["기타대륙"] + ["교포"]
            kto_201901_country_newindex["대륙"] = continents

            kto_201901_country_newindex["관광객비율(%)"] = \
                                                round(kto_201901_country_newindex["관광"] / \
                                                kto_201901_country_newindex["계"] * 100, 1)

            kto_201901_country_newindex["전체비율(%)"] = \
                                        round(kto_201901_country_newindex["관광"] / \
                                                sum(kto_201901_country_newindex["관광"]) * 100, 1)

            df = pd.concat([df, kto_201901_country_newindex], ignore_index=True)
            
        except :
            pass
df
```
