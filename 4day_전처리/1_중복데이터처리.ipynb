```python
# df_temp 데이터를 이용해서 결측치 및 이상치 데이터 처리 연습하기...
# df는 사용하지 않습니다
# 결측치와 이상치 처리 방법은 자유롭게...
df_temp= df.copy()
```

```python
# 이상치 데이터 시각화
import matplotlib as mpl
import matplotlib.pylab as plt
# 이상치 확인하는 그래프 : boxplot(박스플롯)
plt.boxplot(df_temp['나이'])
plt.show()
```

```python
df_col = df_temp.describe().columns
df_col
```

```python
for i in df_col:
    im = df_temp[i].mean()
    df_temp[i].fillna(im, inplace=True)
```

```python
import numpy as np
q1,q3 = np.percentile(df_temp['나이'],[25,75])
iqr = q3-q1
upperbound = q3 + (1.5 * iqr)
lowerbound = q1 - (1.5 * iqr)
```

```python
condition = (df_temp['나이'] > upperbound) | (df_temp['나이'] < lowerbound)
condition.value_counts()
```

```python
df_temp=df_temp[~condition]
```
